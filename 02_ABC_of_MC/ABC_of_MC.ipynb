```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon
```

```python
# Simulate ten thousand draws of a uniformly distributed random variable
np.random.seed(77)
num_draws = 10000
uniform_draws = np.random.uniform(size=num_draws)
```

```python
# Plot the histogram of those draws
plt.hist(uniform_draws, bins=50, density=True)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pass those uniform random draws through the inverse CDF of a normally distributed random variable
normal_draws = norm.ppf(uniform_draws)
```

```python
# Plot the histogram of those new draws
plt.hist(normal_draws, bins=50, density=True)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pass the uniform random draws through the inverse CDF of an exponential random variable with mean 10
mean = 10
exponential_draws = expon.ppf(uniform_draws, scale=mean)
```

```python
# Plot the histogram of those new draws
plt.hist(exponential_draws, bins=50, density=True)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Apply the Probability Integral Transform to the exponential draws
uniform_draws_pit = expon.cdf(exponential_draws, scale=mean)
```

```python
# Plot the histogram of the transformed uniform draws
plt.hist(uniform_draws_pit, bins=50, density=True)
plt.title('Uniform Distribution (PIT)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
