```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm
import seaborn as sns

# Set style for seaborn plots
sns.set(style="whitegrid")
```

```python
# Parameters for the random variable X
mu_X = 10
sigma_X = 5
```

```python
# Number of draws
n = 1000
```

```python
# Draw samples from the normal distribution for X
samples_X = np.random.normal(mu_X, sigma_X, n)
```

```python
# Calculate F(X) using the CDF of X
F_X = norm.cdf(samples_X, loc=mu_X, scale=sigma_X)
```

```python
# Define other CDFs G, H, K with different parameters
G_X = norm.cdf(samples_X, loc=10, scale=10)
H_X = norm.cdf(samples_X, loc=10, scale=2)
K_X = norm.cdf(samples_X, loc=5, scale=5)
L_X = norm.cdf(samples_X, loc=15, scale=5)
```

```python
# Finding the maximum count for the y-axis limit across all histograms
max_count = max(
    np.histogram(samples_X, bins=30)[0].max(),
    np.histogram(F_X, bins=30)[0].max(),
    np.histogram(G_X, bins=30)[0].max(),
    np.histogram(H_X, bins=30)[0].max(),
    np.histogram(K_X, bins=30)[0].max(),
    np.histogram(L_X, bins=30)[0].max(),
)
```

```python
import warnings
warnings.filterwarnings("ignore")

# Create a figure with subplots
fig, axes = plt.subplots(3, 2, figsize=(14, 18))

# Histogram of X
sns.histplot(samples_X, bins=30, color='blue', ax=axes[0, 0], kde=False)
axes[0, 0].set_title("Histogram of X")
axes[0, 0].set_ylim(0, max_count)

# Histogram of F(X)
sns.histplot(F_X, bins=30, color='green', ax=axes[0, 1], kde=False)
axes[0, 1].set_title("Histogram of F(X)")
axes[0, 1].set_ylim(0, max_count)

# Histogram of G(X)
sns.histplot(G_X, bins=30, color='red', ax=axes[1, 0], kde=False)
axes[1, 0].set_title("Histogram of G(X)")
axes[1, 0].set_ylim(0, max_count)

# Histogram of H(X)
sns.histplot(H_X, bins=30, color='purple', ax=axes[1, 1], kde=False)
axes[1, 1].set_title("Histogram of H(X)")
axes[1, 1].set_ylim(0, max_count)

# Histogram of K(X)
sns.histplot(K_X, bins=30, color='orange', ax=axes[2, 0], kde=False)
axes[2, 0].set_title("Histogram of K(X)")
axes[2, 0].set_ylim(0, max_count)

# Histogram of L(X)
sns.histplot(L_X, bins=30, color='brown', ax=axes[2, 1], kde=False)
axes[2, 1].set_title("Histogram of K(X)")
axes[2, 1].set_ylim(0, max_count)

plt.tight_layout()
plt.show()
```
